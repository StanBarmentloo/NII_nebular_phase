```python
import random
import numpy as np
import matplotlib.pyplot as plt
from astropy.convolution import convolve, Gaussian1DKernel, Box1DKernel
from tqdm import tqdm

usesize = 12
plt.rcParams.update({'font.size': usesize})
```

```python
#Set the constants
N = 10**7 # Number of packets in simulation 1E5 default

colorvec=['b','r','k']

tau0vec = np.arange(0, 3.00000001, 0.5)  # For continous simulations (elecronscattering)
v_out_vec = [3000.] #This number does not seem to matter, as the effect has the same relative size at all V

c = 3*10**5
lambda0 = 6316.
```

```python
def translator(x_vals):
    v_out = 3000.
    lambda0 = 6316.
    return x_vals * (v_out/c) * lambda0 + lambda0 
```

```python
for q in tqdm(range(len(v_out_vec))):
    v_out = v_out_vec[q]
    
    peakval = [0]*len(tau0vec)
    peakind = [0]*len(tau0vec)
    xpeak = [0]*len(tau0vec)
    
    wlmin = lambda0*(1-2*v_out/c)
    wlmax = lambda0*(1+6*v_out/c)
    Nbin = 200

    binarr = np.linspace(wlmin,wlmax,Nbin)

    hist = np.zeros(Nbin)
    
    
    for j in tqdm(range(len(tau0vec))):


        tau0 = tau0vec[j]    
        
        print(tau0, v_out)

        hist = np.zeros(Nbin)

        for i in range(0,N):

            wl = lambda0

            x = random.random()

            x = random.random()

            r = x**(1./3)

            mu = 1.-2*random.random()

            escaped = False

            destroyed = False

            while (escaped == False and destroyed == False):

                distedge = -r*mu + np.sqrt((r*mu)**2 - (r**2 - 1**2))

                z = random.random()

                tau_thisphoton = -np.log(1-z) # TEMP righ log?

                distscatt = tau_thisphoton/tau0


                sin_theta = np.sqrt(1.-mu**2)

                if (distscatt < distedge):

                    dr = distscatt

                    r_nextpos = np.sqrt(r**2 + dr**2 - 2*r*dr*(-mu))

                    gamma = 1./np.sqrt(1.-(dr*v_out/c)**2)

                    wl = wl*(1 + v_out/c*dr)*gamma    # Rybicki Lightman eq 4.11 with cos_theta = -1 (homologous flow)

                    mu = 1 - 2*random.random()                  

                else:

                    dr = distedge

                    r_nextpos = 1.

                    sin_theta = r/r_nextpos*sin_theta

                    mu = np.sqrt(1.-sin_theta**2) # mu most be positive reaching surface

                    gamma = 1./np.sqrt(1.-(dr*v_out/c)**2)

                    wl =  wl*(1 + v_out/c*dr)*gamma

                    # In observer frame

                    gamma = 1./np.sqrt(1.-(v_out/c)**2)  

                    wl_escaped = wl*(1. - v_out/c*mu)*gamma  # RybickiLightman eq 4.11


                    # Bin it

                    new = (binarr - wl_escaped)**2

                    ind = new.argmin()

                    hist[ind] = hist[ind] + 1

                    escaped = True

        ynorm = 0.077

        gauss_kernel = Gaussian1DKernel(2)
        hist = convolve(hist, gauss_kernel) #

        #plt.plot((binarr-lambda0)/lambda0/(v_out/c),hist/N/ynorm*Nbin/50,color=colorvec[j]) # TEMP
        plt.plot(binarr,hist/N/ynorm*Nbin/50) # TEMP


        fidsave = open("escatt_data/tau_" +str(int(tau0*10)) + '.csv','w')
        np.savetxt(fidsave,np.c_[(binarr-lambda0)/lambda0/(v_out/c),hist/N/ynorm*Nbin/50/2], delimiter = ',')
        fidsave.close()


        peakval[j] = max(hist[0:int(Nbin/2)]/N/ynorm*Nbin/50)
        peakind[j] = np.argmax(hist[0:int(Nbin/2)]/N/ynorm*Nbin/50)
        xpeak[j] =  (binarr[peakind[j]]-lambda0)/lambda0/(v_out/c)
        print(j, peakval[j], peakind[j], (binarr[peakind[j]]-lambda0)/lambda0/(v_out/c))
```
